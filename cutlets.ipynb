```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
cut = pd.read_csv('Cutlets.csv')
```

```python
cut
```

```python
# h0: Unit A = Unit B
# h1: Unit A != Unit B


observed_table = pd.crosstab(cut['Unit A'],cut['Unit B'])
```

```python
chi2_contingency(observed_table)
```

```python
p_value = chi2_contingency(observed_table)[1]
```

```python
p_value
```

as p value is greater than 0.05 we will go with h0
which is Unit A = Unit B
